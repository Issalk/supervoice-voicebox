```python
import textgrid
import torch
import matplotlib.pyplot as plt
import math
from supervoice.model_vae import VAEModel
from train_config import config

def audio_denormalize(src):
    return (src * config.audio.norm_std) + config.audio.norm_mean

def audio_normalize(src):
    return (src - config.audio.norm_mean) / config.audio.norm_std

model = VAEModel(None)
ch = torch.load("./output/vae_simple.pt", map_location="cpu")
model.load_state_dict(ch["model"])

spec = torch.load("datasets/vctk-prepared/00000068/00055895.pt")
plt.imshow(spec, vmin=-10, vmax=0, origin="lower", aspect="auto")
```

```python
spec2 = model.encode(audio_normalize(spec).unsqueeze(0).transpose(1,2)).transpose(1,2).detach().squeeze(0)
plt.imshow(spec2, origin="lower", aspect="auto")
```

```python
spec3 = audio_denormalize(model.decode(spec2.unsqueeze(0).transpose(1,2)).transpose(1,2).detach().squeeze(0))
plt.imshow(spec3, origin="lower", aspect="auto")
```

```python
# Vocoder
vocoder = torch.hub.load(repo_or_dir='ex3ndr/supervoice-vocoder', model='bigvsan')
def do_vocoder(src):
    with torch.no_grad():
        return vocoder.generate(src)
```

```python
from IPython.display import Audio, display
s = do_vocoder(spec3)
s2 = do_vocoder(spec)
print("source")
display(Audio(data=s2, rate=config.audio.sample_rate))
print("restored")
display(Audio(data=s, rate=config.audio.sample_rate))
```
